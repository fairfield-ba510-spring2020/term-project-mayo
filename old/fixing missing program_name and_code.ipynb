```python
import pandas as pd

df_missing = pd.read_csv(r"CourseCatalog2018_2019.csv")

import pandas as pd

def remove_whitespace(x):
    try:
        # remove spaces inside and outside of string
        x = "".join(x.split())

    except:
        pass
    return x

# Apply remove_whitespace to column only
#df_missing.catalog_id = df_missing.catalog_id.apply(remove_whitespace)
#df.program_code = df.program_code.apply(remove_whitespace)




# Apply to remove_whitespace to entire Dataframe
#df = df.applymap(remove_whitespace)
#print(df_missing)
df_missing.rename(columns = {'Catalog_ID':'catalog_id'}, inplace = True) 
print(df_missing)
```

```python
import re
df_missing.pop('program_code')
df_missing.pop('program_name')
df_missing["program_code"] =df_missing["catalog_id"].str.replace("\d+." , "")
df_missing.program_code = df_missing.program_code.apply(remove_whitespace)
```

```python

df_programs = pd.read_csv("programs_table.csv")
#df_programs= df_programs.applymap(remove_whitespace)
df_programs.program_code = df_programs.program_code.apply(remove_whitespace)
df_programs.isna().sum()
df_programs[df_programs['program_code']== 'FE']
```

```python
df = df_programs[['program_code', 'program_name']].merge(df_missing , left_on=['program_code'], right_on=['program_code'] , how = "right")
df

df.program_name.loc[df.program_code == 'FE'] = 'First Year Experience'
df.program_name.loc[df.program_code == 'ASEN'] = 'American Studies: English'
df.program_name.loc[df.program_code == 'ASAH'] = "American Studies: Art History"
df.program_name.loc[df.program_code == 'ASAH'] = "American Studies: Art History"
df.program_name.loc[df.program_code == 'ASHI'] = "American Studies: History"
df.program_name.loc[df.program_code == 'ASMU'] = "American Studies: Music"
df.program_name.loc[df.program_code == 'ASSO'] = "American Studies: Sociology"

df.program_name.isna().sum()

df[df.program_name.isna() == True]

#df.iloc[1085]
```
